```python
import pandas as pd
from pyod.models.knn import KNN
import lux
import numpy as np
```

```python
df = pd.read_csv("/raw.csv")
df.iloc[:500]
```

```python
###REPLACE WITH MEDIAN 95th Percentile##
df95mn = pd.read_csv('/raw.csv')
# Fit the model
clf = KNN()
clf.fit(df95mn)

# Get the outlier scores
outlier_scores = clf.decision_scores_

# Determine the threshold
threshold = np.percentile(outlier_scores, 95)

# Replace outliers with median
median_value = df95mn.median()
df95mn[outlier_scores > threshold] = median_value

df95mn
```

```python
###REPLACE WITH MEDIAN 99th Percentile##
df99mn = df
# Fit the model
clf = KNN()
clf.fit(df99mn)

# Get the outlier scores
outlier_scores = clf.decision_scores_

# Determine the threshold
threshold = np.percentile(outlier_scores, 99)

# Replace outliers with median
median_value = df99mn.median()
df99mn[outlier_scores > threshold] = median_value

df99mn.iloc[:500]
```

```python
###REPLACE WITH MEAN 95th Percentile##
df95m = df

# Fit the model
clf = KNN()
clf.fit(df95m)

# Get the outlier scores
outlier_scores = clf.decision_scores_

# Determine the threshold
threshold = np.percentile(outlier_scores, 95)

mean_value = df95m.mean()
df95m[outlier_scores > threshold] = mean_value

df95m.iloc[:500]
```
